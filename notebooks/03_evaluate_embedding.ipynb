# Evaluate SCALP-lite Embedding

Score the embedded AnnData file saved by notebook 02.


```python
from scalp_lite.notebook_utils import dataset_config, evaluate_dataset, make_estimator


selected_dataset = "pancreas"
dataset = dataset_config(selected_dataset, embedded=True)

# Evaluation parameters are explicit so this notebook mirrors the flexible style of the sweep notebook.
embedding_key = "X_scalp"
metric_output = "scalp_lite_metrics.csv"

estimator = make_estimator(dataset, n_components=100, random_state=0)
input_path = dataset["input_path"]
csv_path = input_path.parent / metric_output
batch_key = dataset["batch_key"]
label_key = dataset["label_key"]

selected_dataset, input_path, batch_key, label_key, embedding_key
```

```python
scores = evaluate_dataset(estimator, dataset, embedding_key=embedding_key)
scores
```

```python
scores.to_csv(csv_path, index=False)
csv_path
```
