# Parameter Sweep

Load optimized parameters saved by notebook 01 when available; otherwise use dataset defaults. Then sweep one graph parameter at a time and compare paired batch/cell-type embeddings in grids.


```python
import numpy as np
from IPython.display import display

from scalp_lite.notebook_utils import (
    dataset_config,
    load_or_default_params,
    load_preprocessed_data,
    make_estimator,
    run_and_plot_sweep,
)


selected_dataset = "pancreas"
dataset = dataset_config(selected_dataset)
optimized_params = load_or_default_params(selected_dataset, dataset)

max_cells = 6000
random_state = 0
preprocess_overrides = {
    **optimized_params.get("preprocess_params", {}),
    "max_cells": max_cells,
}
estimator_params = optimized_params.get("estimator_params", {})
base_graph_params = optimized_params["graph_params"]

estimator = make_estimator(dataset, random_state=random_state, **estimator_params)
batch_key = dataset["batch_key"]
label_key = dataset["label_key"]
selected_dataset, optimized_params["source"], dataset["input_path"], batch_key, label_key
```

## Data

```python
adata = load_preprocessed_data(estimator, dataset, **preprocess_overrides)
adata
```

## Base Graph Parameters

```python
base_graph_params
```

## Assignment Quantile Sweep

```python
assignment_quantile_results, assignment_quantile_summary, assignment_quantile_fig = run_and_plot_sweep(
    adata=adata,
    estimator=estimator,
    base_graph_params=base_graph_params,
    sweep_parameter="assignment_quantile",
    sweep_values=np.linspace(0.01, 0.1, 5),
    batch_key=batch_key,
    label_key=label_key,
)
display(assignment_quantile_summary)
assignment_quantile_fig
```

## Inter-Batch Edge Sweep

```python
n_inter_edges_results, n_inter_edges_summary, n_inter_edges_fig = run_and_plot_sweep(
    adata=adata,
    estimator=estimator,
    base_graph_params=base_graph_params,
    sweep_parameter="n_inter_edges",
    sweep_values=np.linspace(1, 3, 3).astype(int),
    batch_key=batch_key,
    label_key=label_key,
)
display(n_inter_edges_summary)
n_inter_edges_fig
```

## Neighbor Count Sweep

```python
n_neighbors_results, n_neighbors_summary, n_neighbors_fig = run_and_plot_sweep(
    adata=adata,
    estimator=estimator,
    base_graph_params=base_graph_params,
    sweep_parameter="n_neighbors",
    sweep_values=np.linspace(1, 10, 5).astype(int),
    batch_key=batch_key,
    label_key=label_key,
)
display(n_neighbors_summary)
n_neighbors_fig
```
